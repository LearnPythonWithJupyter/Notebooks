## Getting started

### Download notebooks from [GitHub](www.github.com)
- You can **download all the notebooks** at once:
    - In English at https://github.com/LearnPythonWithJupyter/Notebooks
    - In Español de México at https://github.com/incognia/Notebooks
- Click the green button `< > Code` at the top-right of the GitHub page. Then, click `Download ZIP`
- Unzip the folder on your computer
- Open the first notebook from Jupyter Lab and get started!

### Start coding right away using [notebook.link](https://notebook.link) (no download required!)
- You can **code directly in the browser** at this link: https://notebook.link/github/LearnPythonWithJupyter/Notebooks/tree/main/lab/
- *Note*: It is recommended to **change the notebook theme** to the traditional Jupyter Notebook theme to easily identify the parts of the code (e.g., strings appear in red). To do so, go to `Setting` in the top bar, then `Theme`, and select `Jupyter Lab Light`
- To **save your work**, go to `File`, then `Download` and save the current notebook on your computer
